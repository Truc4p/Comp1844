```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Defining the month names and temperatures for both cities
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
reykjavik_temps = np.array([0.4, 0.5, 1.4, 3.3, 6.1, 9.0, 10.6, 10.3, 7.5, 4.2, 1.4, 0.7])
cairo_temps = np.array([13.6, 15.3, 18.5, 21.7, 25.1, 27.7, 28.3, 28.2, 26.3, 23.1, 18.6, 14.4])

# Creating a DataFrame to display the information
temp_df = pd.DataFrame({
    'Month': months,
    'Reykjavik (째C)': reykjavik_temps,
    'Cairo (째C)': cairo_temps
})

# Printing the table
print(temp_df)
```

```python
# Task 2.1: Line Chart
plt.figure(figsize=(12, 6))
plt.plot(months, reykjavik_temps, marker='o', linestyle='-', label='Reykjavik (째C)')
plt.plot(months, cairo_temps, marker='o', linestyle='-', label='Cairo (째C)')
plt.title('Average Monthly Temperatures: Reykjavik vs Cairo')
plt.xlabel('Month')
plt.ylabel('Temperature (째C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 2.2: Bar Chart
x = np.arange(len(months))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, reykjavik_temps, width, label='Reykjavik (째C)')
bars2 = ax.bar(x + width/2, cairo_temps, width, label='Cairo (째C)')

# Add labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (째C)')
ax.set_title('Average Monthly Temperatures: Reykjavik vs Cairo')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.legend()

plt.show()
```

```python
# Task 2.3: Box Plot Chart
data = [reykjavik_temps, cairo_temps]
fig, ax = plt.subplots(figsize=(8, 6))
ax.boxplot(data, labels=['Reykjavik (째C)', 'Cairo (째C)'])
ax.set_title('Temperature Distribution: Reykjavik vs Cairo')
ax.set_ylabel('Temperature (째C)')
plt.show()
```
