```python
import numpy as np

# Parameters of the normal distribution
Mean = float(5.0) # Mean with capital M is a constant in the code
SD = float(0.8)
# Size of the dataset
Size = int(2000)
# Generate a normal distribution
S = np.random.normal(Mean, SD, Size)
```

```python
import matplotlib.pyplot as plt

# Create a histogram of the data
plt.hist(S, bins=50, alpha=0.75, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution')

# Display the chart
plt.show()
```

```python
# Calculate min, max, and range
min_value = round(np.min(S),2)
max_value = round(np.max(S),2)
range_value = round(np.ptp(S),2)

print(f"Min: {min_value}")
print(f"Max: {max_value}")
print(f"Range: {range_value}")
```

```python
# Plot the histogram
A, X, C = plt.hist(S, bins=20)
# Visualise the plot
plt.show()
```

```python
# Calculate the probability density function
f = (1/(SD*np.sqrt(2*np.pi)))*np.exp(-((X-Mean)**2/(2*SD**2)))
# Plot the probability density function
plt.plot(X, f)
# Visualise the plot
plt.show()
```

```python
# Print the values of A, X, and C
print(f"Counts (A): {A}")
print(f"Bin edges (X): {X}")
print(f"Patches (C): {C}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
Mean = float(5.0)
SD = float(0.8)
# Size of the dataset
Size = int(2000)
# Generate a normal distribution
S = np.random.normal(Mean, SD, Size)

# Create a histogram of the data
A, X, C = plt.hist(S, bins=20)

# Print the values of A, X, and C
print(f"Counts (A): {A}")
print(f"Bin edges (X): {X}")
print(f"Patches (C): {C}")

# Display the chart
plt.show()
```
