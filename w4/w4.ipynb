Task 2

```python
# Check for empty data
print("Null values in each column:")
print(df.isnull().sum())
```

```python
# Fill null values with the mean of all columns except 'ocean_proximity'
numeric_cols = df.select_dtypes(include=['number']).columns
cols_to_fill = numeric_cols.drop('ocean_proximity', errors='ignore')
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].mean())
```

```python
df[cols_to_fill]
```

```python
# Fill NaN values in the 'ocean_proximity' column with the value of the previous row
df['ocean_proximity'] = df['ocean_proximity'].fillna(method='ffill')

df['ocean_proximity']
```

```python
# Combine numeric columns with 'ocean_proximity' column into a new DataFrame df2
df2 = pd.concat([df[cols_to_fill], df['ocean_proximity']], axis=1)
df2
```

```python
# Combine numeric columns with 'ocean_proximity' column into a new DataFrame df2
df2 = pd.concat([df[cols_to_fill], df['ocean_proximity']], axis=1)

# Change value 'NEAR O' to 'NEAR OCEAN' in the 'ocean_proximity' column
df2['ocean_proximity'] = df2['ocean_proximity'].replace('NEAR O', 'NEAR OCEAN')

df2
```

```python
print(df.duplicated())
```

```python
# Check for unnecessary duplicates
print("\nNumber of duplicate rows:", df2.duplicated().sum())

# Remove duplicates
df2 = df2.drop_duplicates()
```

```python
# Change datatype of column "housing_median_age" to integer
df2['housing_median_age'] = df2['housing_median_age'].astype(int)

# Display the DataFrame to verify the change
df2
```

Task 3

```python
# Calculate mean, median, and range of median_house_value
mean = df['median_house_value'].mean()
median = df['median_house_value'].median()
value_range = df['median_house_value'].max() - df['median_house_value'].min()

print(f"Mean of median_house_value: {mean}")
print(f"Median of median_house_value: {median}")
print(f"Range of median_house_value: {value_range}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract median_house_value into a NumPy array
house_values = df['median_house_value'].to_numpy()
```

```python
# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(house_values)
plt.title('Line Chart of Median House Values')
plt.xlabel('Index')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(house_values)
plt.title('Box Plot of Median House Values')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(range(len(house_values)), house_values)
plt.title('Bar Chart of Median House Values')
plt.xlabel('Index')
plt.ylabel('Median House Value')
plt.show()
```
