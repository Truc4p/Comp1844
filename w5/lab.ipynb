```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Task 1: Initialize an array of 20 random values in the interval [0, 100]
np.random.seed(42)  # Setting a seed for reproducibility
random_values = np.random.randint(0, 101, size=20)

# Plot the values using a line chart
plt.figure(figsize=(10, 5))
plt.plot(random_values, marker='o', linestyle='-')
plt.title('Line Chart of Random Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
